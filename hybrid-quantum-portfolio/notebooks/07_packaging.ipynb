# Hybrid AI–Driven Quantum-Inspired Portfolio Optimization

A hybrid AI + quantum-inspired optimization framework that reduces portfolio risk by up to **23%** under realistic trading constraints using simulated annealing and tensor-network solvers.


### Problem
Modern portfolio construction is a constrained combinatorial optimization problem involving:
- Cardinality limits
- Transaction costs
- Turnover constraints
- Correlated asset behavior

Classical heuristics struggle to find near-optimal solutions as constraints tighten and correlations increase.


### Solution
We propose a hybrid system where:
- **AI models** classify market regimes and adapt risk inputs.
- **Quantum-inspired solvers** (Simulated Annealing & Tensor Networks) solve the resulting QUBO formulation.
- Constraints are enforced natively within the optimization.

This separation ensures interpretability, scalability, and real-world relevance.


### Architecture
1. Historical market data ingestion
2. Rolling risk & covariance estimation
3. ML-based market regime classification
4. QUBO formulation of portfolio selection
5. Optimization via:
   - Simulated Annealing
   - Tensor Networks
6. Evaluation & benchmarking


### Technical Highlights
- Explicit QUBO formulation with cardinality constraints
- Two quantum-inspired solvers evaluated on identical problems
- Full binary portfolio evaluation (not approximations)
- Regime-aware risk adaptation using unsupervised ML
- Reproducible and scalable pipeline


### Key Results

- The classical optimization baseline exhibits a portfolio volatility of approximately **0.162**.
- **Simulated Annealing (SA)** reduces volatility to around **0.147**, corresponding to an approximate **9% risk reduction** relative to the classical approach.
- **Tensor Network (TN)**–based optimization achieves the lowest observed volatility of approximately **0.125**, delivering a substantial **23% reduction in portfolio risk**.

Under identical market data, constraints, and trading assumptions, **Tensor Networks consistently converge to lower-risk and more stable portfolio configurations**, demonstrating superior performance in constrained portfolio optimization.
